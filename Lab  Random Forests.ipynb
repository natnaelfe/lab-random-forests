## Instructions
- Apply the Random Forests algorithm but this time only by upscaling the data using SMOTE.
- Note that since SMOTE works on numerical data only, we will first encode the categorical variables in this case.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')

from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
data = pd.read_csv('files_for_lab/Customer-Churn.csv')
data
```

```python
# Cleaning
data.columns = data.columns.str.strip().str.replace(' ', '_').str.lower() 
data.columns
```

```python
# Separate features and target variable
X = data.drop('churn', axis=1)
y = data['churn']
```

```python
# Convert categorical features to one-hot encoded columns
X_encoded = pd.get_dummies(X)
```

```python
# Apply SMOTE for upsampling
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)
```

```python
# Split the upsampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Initialize and fit the Random Forest classifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_rf = rf.predict(X_test)
```

```python
# Calculate accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Classifier Accuracy:", accuracy_rf)
```

Compared to the previous techniques we can see that the Random Forest Classifier Accuracy has the best accuracy! To improve the model we can focus on feature selection or apply other techniques like SelectKBest or ANOVA
